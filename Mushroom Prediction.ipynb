```python
# Collecting Column name in a list.
column_names = ["classes", "cap-shape", "cap-surface", "cap-color", "bruises?", "odor", "gill-attachment",
                "gill-spacing", "gill-size", "gill-color", "stalk-shape", "stalk-root",
                "stalk-surface-above-ring", "stalk-surface-below-ring", "stalk-color-above-ring", "stalk-color-below-ring",
                "veil-type", "veil-color", "ring-number", "ring-type", "spore-print-color",
                "population", "habitat"]
```

```python
# Changing columns names.
df.columns = column_names
df.head()
```

Here is "classes" is dependent variable.
<br>"p" for poisonous and "e" for edible

```python
# Converting e to 1 and p to 0
df["classes"] = df["classes"].replace({"p" : 0, "e" : 1})
df.head()
```

```python
# Checking for nan values
df.info()
```

```python
# Checking unique classes of each variable 
cols = ["cap-shape", "cap-surface", "cap-color", "bruises?", "odor", "gill-attachment",
        "gill-spacing", "gill-size", "gill-color", "stalk-shape", "stalk-root",
        "stalk-surface-above-ring", "stalk-surface-below-ring", "stalk-color-above-ring", "stalk-color-below-ring",
        "veil-type", "veil-color", "ring-number", "ring-type", "spore-print-color",
        "population", "habitat"]
for col in cols:
    print("column:", col)
    print(df[col].unique())
    print("----------------------------------")
```

So, we have "?" in "stalk-root" column.
<br>lets make that as nan.

```python
df["stalk-root"] = df["stalk-root"].replace({"?" : np.nan})
df.head()
```

```python
df["stalk-root"] = df["stalk-root"].fillna(value = "missing")
df.head()
```

#### One - Hot Encoding

```python
X = df.drop(columns = ["classes"])
y = df["classes"]
```

```python
X_encoded = pd.get_dummies(X, drop_first = True)
X_encoded.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, train_size = 0.80)
```

```python
X_train.shape, X_test.shape
```

## Model Building

```python
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
y_pred = lr_model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
plot_confusion_matrix(lr_model, X_test, y_test, display_labels = ["Poisonous", "Edible"])
plt.show()
```

```python
# Testing with cross validation
from sklearn.model_selection import cross_val_score, StratifiedKFold
errors = cross_val_score(lr_model, X_train, y_train, cv = StratifiedKFold(n_splits = 10), scoring = "accuracy")
```

```python
print("Mean Accuracy for 10 Fold CV:",np.mean(errors))
```

```python
# Building final model with all the data
log_model = LogisticRegression()
log_model.fit(X_encoded, y)
```

```python
# Saving the file in .pkl format
import pickle
file = open("mushroom.pkl", mode = "wb")
pickle.dump(log_model, file)
```
